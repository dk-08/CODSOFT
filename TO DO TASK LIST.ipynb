```python
from tkinter import * 
root=Tk()
root.title("TO DO TASK LIST")
root.geometry('600x500')
root.resizable(False,False)
#
lst=[]
def openfile():
    try: 
        global lst  
        with open("file.txt","r" ) as task_list:
            tasks =task_list.readlines()
        for i in tasks:
            if i != '\n':
                lst.append(i)
                lstbox.insert(END,i)
    except:
        file=open("file.txt","w")
        file.close()
#
l1=Label(root,text="TO-DO TASK LIST",font=("arial bold",20),fg="black",bg="pink",highlightbackground="black",highlightthickness=2)
l1.pack()
f=Frame(root,width=500,height=5).pack()
f1=Frame(root,width=400,height=50)
f1.pack()
e1=Entry(f1,width=15,font=("arial",15),bd=2)
e1.pack()
#
item=StringVar()
def addTask():
    item=e1.get()
    e1.delete(0,END)
    if item:
        with open("file.txt","a") as task_list:
            task_list.write(f"\n{item}")
        lst.append(item)
        lstbox.insert(END,item)
b1=Button(f1,text="ADD TO LIST",font=("arial bold",15),bg="pink",fg="black",command=addTask)
b1.pack()
#
ff=Frame(root,width=500,height=5).pack()
f2=Frame(root,width=450,height=200,bg="white",highlightbackground="black",highlightthickness=2)
f2.pack()
lstbox=Listbox(f2,width=50,height=15,font=("arial",10),bg="white",fg="black")
lstbox.pack(side=LEFT,fill=BOTH)
scroll=Scrollbar(f2)
scroll.pack(side=RIGHT,fill=BOTH)
lstbox.config(yscrollcommand=scroll.set)
scroll.config(command=lstbox.yview)
openfile()
#
def delTask():
    global lst 
    item=str(lstbox.get(ANCHOR))
    if item in lst:
        lst.remove(item)
        with open("file.txt","w") as task_list:
            for item in lst:
                task_list.write(item + "\n")
        lstbox.delete(ANCHOR)
b2=Button(root,text="REMOVE TASK",font=("arial bold",15),bg="pink",fg="black",command=delTask)
b2.pack()
root.mainloop()
```
